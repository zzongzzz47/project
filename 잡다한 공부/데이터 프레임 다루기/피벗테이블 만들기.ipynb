```python
df = pd.read_csv('netflix.csv')
df.dtypes
df['release'] = pd.to_datetime(df['release'])
df['year'] = df['release'].dt.year
```

```python
pd.pivot_table(df, values='score', index='year', columns='genre', aggfunc='max')
```
