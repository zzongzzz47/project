```python
import pandas as pd
```

```python
df1 = pd.read_csv('employee.csv')
df2 = pd.read_csv('survey.csv')

pd.merge(df1, df2, on='id')
pd.merge(df1, df2, on = 'id', how='left', suffixes=('_left', '_right'))

```

```python
## 인덱스기준으로 합칠때는 join을 사용한다.

df1 = pd.read_csv('employee.csv')
df2 = pd.read_csv('survey.csv')
df1 = df1.set_index('id')
df2 = df2.set_index('id')
df1.join(df2, rsuffix='_x', how = 'inner')
```
