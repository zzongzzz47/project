```python
import pandas as pd

airbnb_df = pd.read_csv('airbnb.csv')
```

```python
## 결측치 데이터만 확인하고 싶으면 
airbnb_df[airbnb_df.isna().any(axis=1)].head()
```

### 결측값 채우기  
연속형 데이터: 평균값, 중간값  
범주형 데이터: 최빈값

```python
## 결측행 삭제

airbnb_df.dropna()
```

```python
rating_mean = airbnb_df['rating'].mean()
airbnb_df['rating'] = airbnb_df['rating'].fillna(rating_mean)
```

```python
# 중복값 확인 
airbnb_df[airbnb_df.duplicated(subset=['id', 'state'], keep='last')]
```

```python
airbnb_df = airbnb_df.drop_duplicates(subset='id', keep='last')
```

```python
airbnb_df.duplicated(subset='id').sum()
```

```python
q1 = airbnb_df['price'].quantile(0.25)
q3 = airbnb_df['price'].quantile(0.75)

iqr = q3 - q1

lower_limit = q1 - 1.5*iqr
upper_limit = q3 + 1.5*iqr

airbnb_df[~((airbnb_df['price'] < lower_limit)| (airbnb_df['price'] > upper_limit))]
```
