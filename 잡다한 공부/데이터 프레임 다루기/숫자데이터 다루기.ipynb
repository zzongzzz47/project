cut, apply함수 연습

```python
import pandas as pd
```

```python
df = pd.read_csv('patient.csv')
df
```

```python
df['age_group'] = pd.cut(df['age'], 
                         bins = [20, 30, 40, 50, 60, 70],
                         right = False,
                         labels = ['20s', '30s', '40s', '50s', '60s'])
df
```

```python
# apply와 lambda식을 섞어서 사용이 가능하다.

df['height'].apply(lambda x: 39.370079*x)
```
