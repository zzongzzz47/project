```python
import pandas as pd

```

```python
df = pd.read_csv('order.csv')
df['order_time'] = pd.to_datetime(df['order_time'])
df['shipping_time'] = pd.to_datetime(df['shipping_time'])
df = df.dropna()
df['order_time'].dt.dayofweek
```

```python
df = df.set_index('order_time').sort_index()
```

```python
# 부분문자열 인덱싱
df.loc['2015']
df.loc['2015-06']
df.loc['2015-06-10']
```

```python
df['days_for_shipping(actual)'] = df['shipping_time'] - df.index
df.dtypes

df['days_for_shipping(expected)'] = pd.to_timedelta(df['days_for_shipping(expected)'], unit=    )

df['late_shipping'] = df['days_for_shipping(actual)'] > df['days_for_shipping(expected)']
```
