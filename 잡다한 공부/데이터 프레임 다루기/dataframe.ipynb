### 인덱스, 컬럼명, 데이터 타입 설정

```python
loan_df = pd.read_csv('loan.csv')
loan_df = loan_df.set_index('loan_id')
loan_df.reset_index()
loan_df.columns = ['married', 'self_employed', 'applicant_income', 'loan_amount',
                   'property_area', 'loan_status']

new_columns = {'applicant_income':'income', 'loan_amount':'amount'}

loan_df = loan_df.rename(columns = new_columns)
loan_df
```

```python
loan_df['income'] = loan_df['income'].astype('float')
loan_df.dtypes
```

```python
## 

loan_df['self_employed'].unique()
loan_df['self_employed'].value_counts()
loan_df['self_employed'].value_counts(dropna=False)
loan_df['self_employed'].value_counts(dropna=False, normalize=True)
```
